```python
import string
import random

MAX_LIFE = 10
ALPHABET = {alphabet: None for alphabet in string.ascii_lowercase}

# ê²Œì„ì— ì‚¬ìš©ë  ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]

# displayì— í™œìš©ë  hangman pics
hangman_pics_dict = {
    0 : hangman_pics[6],
    1 : hangman_pics[5],
    2 : hangman_pics[4],
    3 : hangman_pics[3],
    4 : hangman_pics[2],
    5 : hangman_pics[1],
    6 : hangman_pics[0],
    7 : hangman_pics[0],
    8 : hangman_pics[0],
    9 : hangman_pics[0],
    10: hangman_pics[0]
}

class HangmanGame:
    def __init__(self):
        # ë§ˆìŒê°™ì•„ì„œëŠ” Stageë¼ëŠ” ê°ì²´ ë”°ë¡œ ë§Œë“¤ì–´ì£¼ê³ ì‹¶ë‹¤.... ì‹œê°„ ë‚¨ìœ¼ë©´ í•´ì•¼ì§€....
        self.stage = None
        self.life = None
        self.word = None

        self.__reset_data()
        self.user_answer = ""
        self.user_answer_list = []

    def __set_user_answer(self, new_user_answer: str) -> None:
        self.user_answer = new_user_answer

    # ë°ì´í„° ì´ˆê¸°í™” ë¡œì§ (ê²Œì„ ì˜¤ë²„ ë˜ê±°ë‚˜, í´ë¦¬ì–´ í›„ ì‹¤í–‰)
    def __reset_data(self) -> None:
        print("ìƒíƒœë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.")
        self.stage = 1
        self.life = MAX_LIFE
        self.word = None

    # ìŠ¤í…Œì´ì§€ ì‹œì‘ ë¡œì§
    def __stage_start(self) -> None:
        # wordsì— ìˆëŠ” ë‹¨ì–´ë¥¼ ëœë¤ìœ¼ë¡œ í•˜ë‚˜ ë½‘ì•„ì„œ ë‹¤ìŒ ìŠ¤í…Œì´ì§€ì˜ ë‹¨ì–´ë¡œ í™œìš©
        print(f"\n\n========================= Stage{self.stage} =========================")
        print(f"Stage{self.stage}. ìƒˆë¡œìš´ ìŠ¤í…Œì´ì§€ê°€ ì‹œì‘ë˜ì–´ ë‹¨ì–´ê°€ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤. ë‹¨ì–´ë¥¼ ë§ì¶°ë³´ì„¸ìš”!!")
        self.word = words.pop(random.randint(0, len(words)-1))

        # ì‚¬ìš©ìì˜ ì •ë‹µì§€ ìƒì„±
        self.__init_user_answer()


    # ìŠ¤í…Œì´ì§€ í´ë¦¬ì–´ í–ˆì„ ë–„ ë¡œì§ (ì¶”ê°€ ë¡œì§ í•„ìš”í•˜ë©´ ì—¬ê¸°ì— ì‘ì„±)
    def __stage_clear(self) -> None:
        # ì‚¬ìš©ìì—ê²Œ í´ë¦¬ì–´ ë³´ìƒ ì œê³µ
        self.__get_stage_clear_benefit()

        # ë‹¤ìŒ ìŠ¤í…Œì´ì§€
        self.stage += 1

    # ìŠ¤í…Œì´ì§€ í´ë¦¬ì–´ í–ˆì„ ë•Œ, ë³´ìƒ
    def __get_stage_clear_benefit(self) -> None:
        if self.life == MAX_LIFE:
            print("ì´ë¯¸ Lifeê°€ ë§Œë•…ì´ì‹œë‹ˆ ìŠ¤í…Œì´ì§€ í´ë¦¬ì–´ ë³´ìƒìœ¼ë¡œ ì €ì˜ ì‚¬ë‘ì„ ì¶©ì „í•´ ë“œë¦¬ê² ìŠµë‹ˆë‹¤.")
            self.__print_life()
        else:
            self.__increase_life()
            print("ìŠ¤í…Œì´ì§€ í´ë¦¬ì–´ ë³´ìƒìœ¼ë¡œ ", end=" ")


    # ì™„ì „íˆ ê²Œì„ í´ë¦¬ì–´
    def __game_clear(self) -> None:
        print("\n\n=+=+=+=+=+=+=+=+=+=+=+=+= Game Clear!! =+=+=+=+=+=+=+=+=+=+=+=+=")
        print(f"ëª¨ë“  ë¬¸ì œë¥¼ í´ë¦¬ì–´í•˜ì…¨ìŠµë‹ˆë‹¤! ë‚¨ì€ ìƒëª… : {self.life}")
        print("ë‹¤ìŒë²ˆì—” ë” ë§ì€ ìƒëª…ì„ ë‚¨ê¸¸ ìˆ˜ ìˆë„ë¡ ë…¸ë ¥í•´ë³´ì„¸ìš”!")
        print(" - FIN -  CopyRight. Bbong Company")
        print("=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=\n\n")
        self.__quit_game()

    # ì™„ì „íˆ ê²Œì„ ì˜¤ë²„
    def __game_over(self) -> None:
        print(f"ê²Œì„ì„ í´ë¦¬ì–´í•˜ì§€ ëª»í–ˆìŠµë‹ˆë‹¤... ì •ë‹µì€ {self.word} ì´ì—ˆìŠµë‹ˆë‹¤..!")
        self.__quit_game()

    # ê²Œì„ ì¢…ë£Œ
    def __quit_game(self) -> None:
        print("ê²Œì„ì´ ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
        self.__reset_data()


    # ë§ˆì§€ë§‰ ìŠ¤í…Œì´ì§€ì¸ì§€ ì²´í¬í•©ë‹ˆë‹¤.    
    def __is_last_stage(self) -> None:
        return len(words) == 0

    # ì •ë‹µì¸ì§€ ì²´í¬í•©ë‹ˆë‹¤.
    def __is_stage_clear(self) -> bool:
        return self.word == self.user_answer

    # ì‚´ì•˜ëŠ”ì§€ ì²´í¬í•©ë‹ˆë‹¤.
    def __is_alive(self) -> bool:
        return self.life > 0
    
    # ì‚¬ìš©ìê°€ ì…ë ¥í•œ ë‹¨ì–´ê°€ ì •ë‹µì— í¬í•¨ë˜ì–´ìˆëŠ”ì§€ í™•ì¸í•©ë‹ˆë‹¤.
    def __check_user_input(self, user_input: str) -> bool:
        return user_input in self.word

    # ìˆ¨ê²¨ì§„ ì •ë‹µì„ ì—´ê²Œë©ë‹ˆë‹¤.
    def __open_word(self, user_input: str) -> str:
        def get_covered_index_list(user_input: str) -> list[int]:
            covered_index_list = []

            for idx, char in enumerate(self.word):
                if char == user_input:
                    covered_index_list.append(idx)

            return covered_index_list

        covered_index_list = get_covered_index_list(user_input=user_input)

        user_answer_char_list = list(self.user_answer)
        for idx in covered_index_list:
            user_answer_char_list[idx] = user_input

        self.__set_user_answer(new_user_answer="".join(user_answer_char_list))

    # ì‚¬ìš©ìì˜ ì‘ë‹µì„ ì–»ìŠµë‹ˆë‹¤.
    def __get_user_input(self) -> str:
        while True:
            user_input = input("ì›í•˜ëŠ” ì†Œë¬¸ì ì•ŒíŒŒë²³ì„ 1ê¸€ìë§Œ ì…ë ¥í•˜ì„¸ìš” (ì¢…ë£Œë¥¼ ì›í•œë‹¤ë©´ 'ì¢…ë£Œ'ë¥¼ ì…ë ¥í•˜ì„¸ìš”) : ")
            
            if user_input == "ì¢…ë£Œ":
                return user_input

            if self.__validate_user_input(user_input=user_input):
                # ì •ë‹µ ì¤‘ë³µ ì…ë ¥ ë°©ì§€ë¥¼ ìœ„í•´ ì…ë ¥ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
                self.user_answer_list.append(user_input)
                return user_input
        
    def __print_user_answer(self) -> None:
        user_answer_for_print = " ".join(self.user_answer) 
        print(f"\ní˜„ì¬ ì‚¬ìš©ìì˜ ì •ë‹µ í˜„í™© : {user_answer_for_print}")

    def __validate_user_input(self, user_input: str) -> bool:
        # ê¸¸ì´ê°€ 1ê¸€ìê°€ ì•„ë‹Œ ê²½ìš° (ë‚˜ì¤‘ì— ë¡œì§ ë§ì•„ì§€ë©´ ë³„ë„ ë©”ì„œë“œë¡œ ë¶„ë¦¬)
        if len(user_input) != 1:
            print("ì†Œë¬¸ì ì•ŒíŒŒë²³ 1ê¸€ìë§Œ ì…ë ¥í•´ ì£¼ì„¸ìš”")
            return False
        
        # ë‹¨ì–´ê°€ ì†Œë¬¸ì ì•ŒíŒŒë²³ì´ ì•„ë‹Œ ê²½ìš°
        if user_input not in ALPHABET:
            print("ì†Œë¬¸ì ì•ŒíŒŒë²³ 1ê¸€ìë§Œ ì…ë ¥í•´ ì£¼ì„¸ìš”")
            return False
        
        if user_input in self.user_answer_list:
            print("ì´ë¯¸ ì…ë ¥í•˜ì…¨ë˜ ì‘ë‹µì…ë‹ˆë‹¤. ë‹¤ì‹œí•œë²ˆ ìƒê°í•´ë³´ì„¸ìš”")
            return False
        
        return True

    def __reduce_life(self, x: int = 1) -> None:
        self.life -= x
        print(f"Lifeê°€ {x}ë§Œí¼ ê°ì†Œí–ˆìŠµë‹ˆë‹¤.")
        self.__print_life()

    def __increase_life(self, x: int = 1) -> None:
        self.life += x
        print(f"Lifeê°€ {x}ë§Œí¼ ì¶”ê°€ ë˜ì—ˆìŠµë‹ˆë‹¤")
        self.__print_life()

    def __print_life(self) -> None:
        print(f"í˜„ì¬ Life : [{self.life}/{MAX_LIFE}]")
        self.display()

    # ì‚¬ìš©ìì˜ ì •ë‹µì§€ë¥¼ ìƒì„± ë° ì¤‘ë³µ ë°©ì§€ìš© ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤
    def __init_user_answer(self) -> None:
        self.__set_user_answer("_" * len(self.word))
        self.user_answer_list.clear()



    def display(self) -> None:
        hangman_pic = hangman_pics_dict.get(self.life)
        print(hangman_pic)


    def play(self) -> None:
        # ì§€ê¸ˆì´ ë§ˆì§€ë§‰ ìŠ¤í…Œì´ì§€ê°€ ì•„ë‹ˆë¼ë©´ (ìŠ¤í…Œì´ì§€ ë°˜ë³µë¬¸)
        while not self.__is_last_stage():
            # ìŠ¤í…Œì´ì§€ ì‹œì‘
            self.__stage_start()
    
            # ë°˜ë³µí•´ì„œ ì •ë‹µ ì—¬ë¶€ í™•ì¸
            while not self.__is_stage_clear():
                self.__print_user_answer()
                # ì‚¬ìš©ìì˜ ì…ë ¥ ë°ì´í„° ì œê³µ
                user_input = self.__get_user_input()

                # ì‚¬ìš©ìê°€ ì¤‘ê°„ì— ì¢…ë£Œë¥¼ ì…ë ¥í–ˆë‹¤ë©´, ê²Œì„ ì¢…ë£Œ
                if user_input == "ì¢…ë£Œ":
                    self.__quit_game()
                    return

                # ì •ë‹µì¸ì§€ í™•ì¸
                if self.__check_user_input(user_input=user_input):
                    # ì •ë‹µì¼ ê²½ìš° ì‚¬ìš©ì ì •ë‹µì§€ì— ë‹¨ì–´ ê³µê°œ
                    self.__open_word(user_input=user_input)
                    print("ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨êµ°ìš”. ë” í˜ë‚´ë´ìš”")
                else:
                    # ì˜¤ë‹µì´ë¼ë©´ ìƒëª…ì´ ê¹ì„
                    self.__reduce_life()

                    # ê²Œì„ ì§€ì† ê°€ëŠ¥í•œì§€ í™•ì¸
                    if self.__is_alive():
                        print(f"í‹€ë ¸ìŠµë‹ˆë‹¤. ë‚¨ì€ ì‹œë„ íšŸìˆ˜ : {self.life}")
                    else:
                        # ì§€ì†ì´ ë¶ˆê°€ëŠ¥í•˜ë‹¤ë©´ ê²Œì„ ì˜¤ë²„
                        self.__game_over()
                        return

            # ì •ë‹µì¼ ê²½ìš° ìŠ¤í…Œì´ì§€ í´ë¦¬ì–´ íŒì • í›„ ë‹¤ìŒ ìŠ¤í…Œì´ì§€
            self.__stage_clear()

        # ë” ì´ìƒ ì§€ì†í•  ë‹¨ì–´ê°€ ì—†ì„ ê²½ìš° ê²Œì„ í´ë¦¬ì–´
        self.__game_clear()
```
