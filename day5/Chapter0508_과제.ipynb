[문제1]

변수명 게임에 들어 있는 값 중 '홀'만 출력되도록 코드를 작성해 주세요

게임 = "홀짝홀짝홀짝"

출력 예시 : "홀홀홀"


```python
"""
입력한 단어가 "홀"과 일치하는지 여부를 판단합니다.

@Param
input : 체크할 단어 하나를 입력합니다.

@return
bool -> 체크한 단어가 홀인지 True, False로 반환합니다.
"""
def check_홀(input: str) -> bool:
    if (0 >= len(input)) or (1 < len(input)):
        raise ValueError("입력 값의 길이가 맞지 않습니다. 딱 한글자만 넣어주세요")

    return input == '홀'

game = "홀짝홀짝홀짝"
print_result = ''

for i in game:
    if check_홀(i):
        print_result += i

print(print_result)
```

[문제3]

창의성, 시장성, 안정성, 수익성에 대한 각각의 평가 중 아래 기준에 부합하지 못하면 창업지원을 받을 수 없습니다.

- 수익성 60점 이상
- 창의성과 시장성 50점 이상
- 시장성과 수익성을 더한 값의 절반이 안정성 보다 높아야 함

다음 점수가 입력되었을때 창업 지원이 가능한지 여부를 True 또는 False로 출력해주세요

---

창의성 : 62
시장성 : 71
안정성 : 68
수익성 : 66

```python
# 성공 기준
# creativity, marketability, stability, profitability = 62, 71, 68, 66

# 실패 기준
creativity, marketability, stability, profitability = 62, 71, 68, 46


"""
창업의 기준이 되는 4가지 값(창의성, 시장성, 안정성, 수익성)을 활용하여
창업 가능 여부를 판단합니다.

@Param
creativity : 창의성 값을 입력합니다.
marketability : 시장성 값을 입력합니다.
stability : 안정성 값을 입력합니다.
profitability : 수익성 값을 입력합니다.

@return
bool -> 창업 가능 여부 결과를 True, False로 반환합니다.
"""
def check_market_fit(
        creativity: int, 
        marketability: int, 
        stability: int, 
        profitability: int) -> bool:
    
    if profitability < 60:
        print(f"수익성이 {profitability}점으로 기준에 충족하지 못합니다.")
        return False
    
    if (marketability < 50) or (creativity < 50):
        print(f"창의성과 시장성 중 50점을 충족하지 못하는 항목이 존재합니다.")
        return False
    
    calculate_m_p = (marketability + profitability) / 2

    if (calculate_m_p <= stability):
        print(f"시장성과 수익성을 활용한 연산 결과가 {calculate_m_p}점으로 안정성 점수를 넘지 못합니다.")
        return False
    
    return True

market_fit_result = check_market_fit(
    creativity=creativity, 
    marketability=marketability, 
    stability=stability, 
    profitability=profitability
    )

print(f"창업 지원 가능 여부 결과는 : {market_fit_result} 입니다.")
```
